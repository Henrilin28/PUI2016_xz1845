```python
from __future__ import print_function
__author__='Xinshi Zheng'

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
from scipy.optimize import curve_fit, minimize

%matplotlib inline
```

```python
np.random.seed()
```

```python
mean = 100
```

```python
size = np.linspace(11, 1999,100).astype(int)
```

```python
ls = []
```

```python
for i in size:
    sample_mean = np.random.chisquare(mean, size = i).mean()
    ls.append(sample_mean)
    plt.plot(i, sample_mean, '.')
    plt.xlabel('Sample Size')
    plt.ylabel('Sample Mean')
    plt.title('Chisquare')
```

```python
for i in size:
    sample_mean = np.random.normal(loc = mean, size = i).mean()
    ls.append(sample_mean)
    plt.plot(i, sample_mean, '.')
    plt.xlabel('Sample Size')
    plt.ylabel('Sample Mean')
    plt.title('Normal')
```

```python
for i in size:
    sample_mean = np.random.poisson(lam = mean, size = i).mean()
    ls.append(sample_mean)
    plt.plot(i, sample_mean, '.')
    plt.xlabel('Sample Size')
    plt.ylabel('Sample Mean')
    plt.title('Poisson')
```

```python
for i in size:
    sample_mean = np.random.binomial(n = 200, p = 0.5, size = i).mean()
    ls.append(sample_mean)
    plt.plot(i, sample_mean, '.')
    plt.xlabel('Sample Size')
    plt.ylabel('Sample Mean')
    plt.title('Binomial')
```

```python
for i in size:
    sample_mean = np.random.logistic(loc = mean, size = i).mean()
    ls.append(sample_mean)
    plt.plot(i, sample_mean, '.')
    plt.xlabel('Sample Size')
    plt.ylabel('Sample Mean')
    plt.title('Logistic')
```

```python
plt.hist(ls, bins=30)
plt.xlabel('Sample Mean')
plt.ylabel('N')
plt.show()
```

```python
plt.plot(ls, '.')
plt.xlabel('Iteration')
plt.ylabel('Sample Mean')
plt.show()
```

```python
plt.hist(ls, bins=30, normed = True)

mean_ls = np.mean(ls)
var_ls = np.var(ls)
sigma_ls = np.sqrt(var_ls)
x_ls = np.linspace(min(ls), max(ls),100)
plt.plot(x_ls,mlab.normpdf(x_ls,mean_ls,sigma_ls))

plt.xlabel('Sample Mean')
plt.ylabel('N')
plt.title('Gaussian Curve Fitting')
plt.show()
```
