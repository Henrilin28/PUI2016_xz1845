```python
# Write a function to fit a line through log data and calculate chi-square.
def calChisquare(x, y):
    X = sm.add_constant(np.log10(x))
    linmodel = sm.OLS(np.log10(y), X, missing='drop').fit()
    Y = linmodel.predict()
    chisq = ((Y - np.log10(y))**2 / (np.log10(y)/2)).sum()
    return ('The Chi-Square is {}'.format(chisq))
```

```python
calChisquare(bblnrgdata_cut.UnitsTotal, bblnrgdata_cut.energy_total)
```

```python
calChisquare(bblnrgdata_cut.energy_total, bblnrgdata_cut.UnitsTotal)
```

# From the Chi-square results it may be better to switch independent and dependent variables

```python
# Linear regression model for the log data set
X = sm.add_constant(np.log10(bblnrgdata_cut.energy_total))
linmodel = sm.OLS(np.log10(bblnrgdata_cut.UnitsTotal), X, missing='drop').fit()
print(linmodel.summary())
```

```python
# Create new columns in the dataframe with log values
bblnrgdata_cut['log_UnitsTotals'] = np.log10(bblnrgdata_cut.UnitsTotal)
bblnrgdata_cut['log_EnergyTotals'] = np.log10(bblnrgdata_cut.energy_total)
```

```python
# Make log plot of energy use angainst units number
plt.figure(figsize=(10,10))
plt.scatter(bblnrgdata_cut.log_EnergyTotals, 
            bblnrgdata_cut.log_UnitsTotals, 
            c='r', alpha=0.6, s=40, label='Observation')
x = np.linspace(3, 9, 7)
plt.plot(x, linmodel.predict(sm.add_constant(x)), c='b', label='linear regression fit')
plt.xlabel('Total Energy Consumption (kBtu) (log)')
plt.ylabel('Number of Units (log)')
plt.title('Log Plot of Unit Size VS Energy Consumption')
plt.legend(loc='best')
plt.show()
```

# Figure 6: Log Plot of Number of Units against Total Energy Consumption with Linear Regression Fit

```python
# Create a new column in the dataframe for 2nd degee polynomial fitting
bblnrgdata_cut['squared_log_EnergyTotals'] = bblnrgdata_cut.log_EnergyTotals ** 2

# Fit a 2nd degee polynomial
polynomial_2nd = smf.ols(formula = 'log_UnitsTotals ~ squared_log_EnergyTotals + log_EnergyTotals', data=bblnrgdata_cut).fit()
```

```python
print(polynomial_2nd.summary())
```

```python
# Make log plot of energy use angainst units number
plt.figure(figsize=(10,10))
plt.scatter(bblnrgdata_cut.log_EnergyTotals, 
            bblnrgdata_cut.log_UnitsTotals, 
            c='r', alpha=0.6, s=40, label='Observation')
x = np.linspace(3, 9, 100)
plt.plot(x, linmodel.predict(sm.add_constant(x)), c='b', label='linear regression fit')
plt.plot(x, polynomial_2nd.predict(pd.DataFrame({'log_EnergyTotals': x, 'squared_log_EnergyTotals': x**2})), 
         c='k', label='2nd polynomial fit')
plt.xlabel('Total Energy Consumption (kBtu) (log)')
plt.ylabel('Number of Units (log)')
plt.title('Log Plot of Unit Size VS Energy Consumption')
plt.legend(loc='best')
plt.show()
```

# Figure 7: Log Plot of Number of Units against Total Energy Consumption with Linear Regression Fit and 2nd Polynomial Fit

```python
# Assume a significance level of 0.05 that linear model is preferable to the 2nd polynomial model 
# using the likelihood ratio test. The corresponding alpha = 3.84 from the chi-square table with DOF of 1.

# Calculate the Likelihood ratio test
LR = -2 * (linmodel.llf - polynomial_2nd.llf)

if LR > 3.84:
    print('LR is {}. Therefore, the 2nd polynomial model is preferable to the linear model.'.format(LR))
else:
    print('LR is {}. Therefore, the linear model is preferable to the 2nd polynomial model.'.format(LR))
```

```python
# Use compare_lr_test() method to confirm
polynomial_2nd.compare_lr_test(linmodel)
```

# The second order polynomial model is preferable.

```python
# Define the residual minimization function
def residual_min(slope, intercept):
    return(((bblnrgdata_cut.log_UnitsTotals - 
           (bblnrgdata_cut.log_EnergyTotals * slope + intercept))**2 / (bblnrgdata_cut.log_UnitsTotals / 2)) ** 2).sum()
```

```python
# Populate the space with different slopes and intercepts combinations
intercept = np.arange(-100, 100, 1)
slope = np.arange(-50.0, 50.0, 0.5)
Lsurface = np.zeros((len(slope), len(intercept)))
for i, a in enumerate(slope):
    for j, b in enumerate(intercept):
        Lsurface[i][j] = residual_min(a, b)
```

```python
plt.figure(figsize=(10,10))
plt.title ("log likelihood surface", fontsize = 22)
plt.imshow(np.log(Lsurface), extent = [-50,50,100,-100], aspect=0.5, cmap='bone')
plt.xlabel('slope', fontsize = 22)
plt.ylabel('intercept', fontsize = 22)
plt.colorbar()
plt.show()
```
